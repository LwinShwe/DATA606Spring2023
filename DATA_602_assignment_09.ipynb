## **Bar Plot**

A barplot shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value.

```python
px.bar(df,color="total_bill", x="tip", y="day", hover_name="time", height=600)
```

```python
df = px.data.tips()
px.bar(df,color="smoker", x="sex", y="total_bill", hover_name="time", title= "Tips Amount")
```

```python
px.bar(df,color="day", x="sex", y="total_bill", hover_name="time", title= "Tips Amount")
```

## **Scatter Plot**

Scatter plots are used to check the relationship between the variables and the distribution of the data. A scatterplot displays the relationship between 2 numeric variables. For each data point, the value of its first variable is represented on the X axis, the second on the Y axis

```python
df = px.data.tips()
px.scatter(df,color="smoker", y="tip", x="total_bill", hover_name="smoker", facet_col="sex", facet_row="time")
```

```python
#adding trend lines
df=px.data.tips()
px.scatter(df,color="smoker", y="tip", x="total_bill", hover_name="smoker", facet_col="sex", facet_row="time", trendline="ols")
```

```python
#exploring vs. day
df=px.data.tips()
px.scatter(df,color="smoker", y="tip", x="total_bill", hover_name="smoker", facet_col="day", facet_row="time")
```

```python
df=px.data.tips()
px.scatter(df,color="smoker", y="tip", x="total_bill", hover_name="smoker", facet_col="day", facet_row="time", trendline="ols")
```

# **Histogram**

A histogram takes as input a numeric variable only. The variable is cut into several bins, and the number of observation per bin is represented by the height of the bar. It is possible to represent the distribution of several variable on the same axis using this technique.

```python
#Exploring distrubution of tip
px.histogram(df,x="total_bill", y="tip", color="day", hover_name="time", marginal="rug", facet_col="day")
```

# **Boxplot**

A boxplot gives a nice summary of one or several numeric variables. The line that divides the box into 2 parts represents the median of the data

```python
df = px.data.tips()
px.box(df,x="smoker", y="tip", points="outliers", color="smoker")
```

```python
#Styled Box Plot
df = px.data.tips()
px.box(df,x="smoker", y="tip", points="outliers", color="smoker", notched =True, title="Box Plot of Total_Tip Amount", hover_data=["day"])
```
