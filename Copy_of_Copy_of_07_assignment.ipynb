```
# Group your dataset by one column, and get the mean, min, and max values by group.
# Groupby(), agg() or .apply()

# Import pandas 
import pandas as pd

# Use pandas to read nyc weather 
weather_df = pd.read_csv('/content/sample_data/NYC_Weather_2016_2022.csv')

grouped = weather_df.groupby("time")
grouped = weather_df.groupby(["time", "windspeed_10m (km/h)"])
grouped.sum()
```

```
# group by 'Temperature_2m', get mean, min, and max value of 'winddirection_10m (°)'for each value of Temperature.
grouped_single = weather_df.groupby('temperature_2m (°C)').agg({'winddirection_10m (°)': ['mean', 'min', 'max']})

print(grouped_single)
```

```
# Group your dataset by two columns and then sort the aggregated results within the groups
grouped_multiple = weather_df.groupby(['temperature_2m (°C)', 'cloudcover_high (%)']).agg({'winddirection_10m (°)': ['mean', 'min', 'max']})
grouped_multiple.columns = ['winddirection_10m (°)_mean', 'winddirection_10m (°)_min', 'winddirection_10m (°)_max']
grouped_multiple = grouped_multiple.reset_index()
print(grouped_multiple)
```
